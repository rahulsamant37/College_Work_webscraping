```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
webpage = requests.get('https://docs.google.com/spreadsheets/u/0/d/12upB23MbwnO_qvSVVDGrr65hBwtnxnr0/htmlview#gid=398746635').text
```

```python
soup=BeautifulSoup(webpage,'lxml')
```

```python
len(soup.find_all('tr'))
```

```python
name=[]
contact=[]
specialization=[]
for index,i in enumerate(soup.find_all('tr')):
  if index>3:
    td_element=i.find_all('td')
    if td_element[0]:
      name.append(td_element[0].text)
    if td_element[1]:
      contact.append(td_element[1].text)
    if td_element[2]:
      specialization.append(td_element[2].text)
```

```python
d={'Name':name,'Contact':contact,'Specialization':specialization}
```

```python
df=pd.DataFrame(d)
df.index = range(1, len(df) + 1)
```

```python
# Function to extract website and email
def extract_contact_info(contact_str):
    if not isinstance(contact_str, str):
        # If the contact_str is not a string, return NaN values for both
        return pd.Series([None, None], index=['Website', 'Email'])

    website = None
    email = None

    # Handle cases where there's no space between website and email
    if 'Website:' in contact_str:
        website_start = contact_str.find('Website: ') + len('Website: ')
        website_end = contact_str.find(' Email:', website_start)
        if website_end == -1:
            website_end = contact_str.find('Email:', website_start)
        if website_end == -1:
            website_end = len(contact_str)
        website = contact_str[website_start:website_end].strip()

    if 'Email:' in contact_str:
        email_start = contact_str.find('Email:') + len('Email:')
        email = contact_str[email_start:].strip()

    return pd.Series([website, email], index=['Website', 'Email'])

# Apply the function to the 'Contact' column
d[['Website', 'Email']] = d['Contact'].apply(extract_contact_info)


# Display the DataFrame to see the result
print(d)
```

```python
d = d.drop(columns=['Name'])
```
